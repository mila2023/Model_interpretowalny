```python
import pandas as pd
import numpy as np
from statistics import mean, stdev
from sklearn import preprocessing
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn import linear_model
from sklearn import datasets
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("zbiór_10.csv")

# rozkladamy dane na wejsciowe i wyjsciowe i dropujemy skurwysyna szczegolna forma wlasnosci, bo nic nie wnosi = default 117
X = data.drop(columns=["default", "szczegolnaFormaPrawna_Symbol"])
y = data["default"]

model = LogisticRegression(max_iter=1000)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')

# for train_id, test_id in folds.split(X):
#     print("Train indices: ", train_id, "Test indices: ", test_id)

print("Cross-Validation Scores:", scores)
print("Mean accuracy:", scores.mean())

```
